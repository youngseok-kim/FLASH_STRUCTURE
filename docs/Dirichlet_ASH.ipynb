Suppose we have independent observation $x_j$ from Categorical$(\pi_j)$ for $j = 1,\cdots, n$ where $x_j$'s and $\pi_j$'s are of length $r$.

$$
\log p(x_j|\pi_{j}) = a(x_j) + \sum_{l = 1}^r x_{jl} \log \pi_{jl}
$$

and $\pi_j$'s are from a mixture of symmetric Dirichlet distributions

$$
\log p(\pi_j|t_k) = h(t_k,t_k,\cdots,t_k) + t_k \sum_l \log \pi_{jl}
$$

Note that it is Dirichlet$(t_k+1,\cdots,t_k+1)$.

Let $t_k = c^{-k+1} - 1$ for $k = 1,...,m$ for some $c > 1$.

For a fixed $t_k$, a marginal

$$
p(x_j|t_k) = \int p(x_j|\pi_j)p(\pi_j|t_k) d\pi_j
$$

is proportional to $a(x_j)h(t_k,\cdots,t_k)h(x_{j1} + t_k,\cdots,x_{jr} + t_k)$ and it is easy to check that

$$
h(\alpha_1,\cdots,\alpha_r) = -\log B(\alpha_1+1,\cdots,\alpha_r+1) = \log \Gamma\left( l + \sum_l \alpha_l \right) -\sum_{l} \log \Gamma(\alpha_l + 1)
$$

```julia
# I think we can make a general ash if we specify h_diff
function h_diff(x,t)
    n,r = size(x); m = length(t);
    
    function h(α)
        return -sum(lgamma.(α),2) .+ lgamma.(sum(α,2))
    end
    
    out = h(repeat(t+1, inner = (n,r))) .- h(repmat(x,m,1) + repeat(t+1, inner = (n,r)));
    return reshape(out, n, length(t))
end

function ash_dirichlet(x::Array{Float64,2}, h_diff::Function;
                       t::Array{Float64,1} = 1.5.^(-(1:20) + 1) - 1)
    # set sizes
    n,r = size(x);
    m = length(t);
    
    # contruct a likelihood function
    loglik = h_diff(x,t)
    lik = exp.(loglik .- maximum(loglik,2));
    
    # solve mixsqp
    out = mixsqp(lik, verbose = false)["x"];
    ind = find(out .> 0);
    
    # the posterior is a mixture of Dirichlet distributions
    # Dirichlet(x_{j1} + t_k + 1, ..., x_{jr} + t_k + 1)
    
    # component posterior probability
    L = lik[:,ind];
    cpp = L .* out[ind]';
    cpp = cpp ./ sum(cpp,2);
    
    # component posterior mean
    s = length(ind);
    cpm = repmat(x,s,1) + repeat(t[ind]+1, inner = (n,r))
    cpm = cpm./sum(cpm,2);

    # posterior mean
    pm = repmat(speye(1000),1,5) * (cpm .* cpp[:]);
    
    return Dict([
                (:p, out), (:pp, out[ind]), (:L, L), (:t, t[ind]), (:lik, lik),
                (:cpp, cpp), (:cpm, cpm), (:pm, pm)
                ])
end
```

```julia
include("../src/mixsqp.jl")

out = ash_dirichlet(x * 100, h_diff)
```
